## **Model Building**

**Understanding the Objective**

Goal: Predict whether a patient has diabetic retinopathy (1) or not (0), using features like blood pressure, cholesterol, and age.

> Type of Problem:

This is a binary classification problem because the target (prognosis) has only two possible values:

1 = Retinopathy
0 = No Retinopathy


```python
# Reload the data properly (with semicolon separator)
df = pd.read_csv("pronostico_dataset (1).csv", delimiter=';')

# Check unique values in 'prognosis' BEFORE encoding
print(df['prognosis'].unique())
print(df['prognosis'].value_counts())
```

**Data Preparation (after cleaning)**

cleaned and scaled the features

Target = 'prognosis'

```python
# Drop ID column
df = df.drop(columns=['ID'])
```

```python
# Encode target variable
le = LabelEncoder()
df['prognosis'] = le.fit_transform(df['prognosis'])  # 'no_retinopathy' → 0, 'retinopathy' → 1
```

```python
# Confirm encoding
print(df['prognosis'].value_counts())  # Should show 1s and 0s
```

**Train-Test Split**

We split the data into:

Training set → To learn

Testing set → To evaluate

```python
# Split features and target
X = df.drop(columns=['prognosis'])
y = df['prognosis']
```

```python
# Train-test split (stratify for class balance)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Convert to DataFrame for easy viewing
pd.DataFrame(X_train_scaled, columns=X.columns).head()
```

```python
# Create interpretation table

data = {
    "Feature": ["age", "systolic_bp", "diastolic_bp", "cholesterol"],
    "Scaled Value": ["-0.67 to 0.53", "-1.80 to 2.53", "-0.76 to 1.15", "-1.70 to 0.83"],
    "Interpretation": [
        "- Values < 0 = younger than average\n- Values > 0 = older than average",
        "- Big range → some have very low/high blood pressure",
        "- Normalized around 0 (low < 0, high > 0)",
        "- Below/above average cholesterol levels"
    ]
}

# Convert to DataFrame
interpretation_df = pd.DataFrame(data)

# Display
interpretation_df
```

**Logistic Regression**

This is the simplest ML classifier for binary problems.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score

# Step 1: Train the model
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

# Step 2: Predict
lr_preds = lr_model.predict(X_test)

# Step 3: Evaluate
print("Confusion Matrix:\n", confusion_matrix(y_test, lr_preds))
print("Classification Report:\n", classification_report(y_test, lr_preds))
print("AUC Score:", roc_auc_score(y_test, lr_preds))
```

### Confusion Matrix Explanation

|--------------------|--------|---------|

| False Positives    | 126    | Predicted DR but actually healthy |

| True Positives     | 464    | Correctly predicted DR cases |

---

### Classification Report Summary

- **Precision (Class 1)** = 0.79 → 79% of predicted DR cases are correct
- **Recall (Class 1)** = 0.75 → 75% of actual DR cases are detected
- **F1 Score** = 0.77 → Good balance between precision and recall
- **Overall Accuracy** = 77% → 3 out of 4 predictions are correct

---

### AUC Score

- **AUC = 0.768**
- Measures how well the model distinguishes between DR and non-DR

**Random Forest**

We’ll start with a Random Forest Classifier (good for structured data):

```python
from sklearn.ensemble import RandomForestClassifier

# Choose a Model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Model Fit
rf_model.fit(X_train, y_train)

# Test the Model
rf_preds = rf_model.predict(X_test)

print("Confusion Matrix:\n", confusion_matrix(y_test, rf_preds))
print("Classification Report:\n", classification_report(y_test, rf_preds))
print("AUC Score:", roc_auc_score(y_test, rf_preds))
```

Precision: When the model said "this is class 1", how often was it right?
→ Around 73% of the time, it was correct.

Recall: Out of all the real class 1s, how many did the model find?
→ It caught 74% of them.

F1-score: A mix of both precision and recall. Think of it as the overall score for each class.

Accuracy: It got 73% of all predictions correct.

**Visualize Performance – ROC Curve**

The ROC curve shows the balance between sensitivity and specificity.

**The ROC Curve**- 
The ROC Curve is a graph that shows how well a model can tell the difference between two classes (like 0 and 1). It compares:

- True Positive Rate (TPR) – how many real positives it correctly predicted

- False Positive Rate (FPR) – how many negatives it incorrectly predicted as positive

The higher the curve and the bigger the area under it (AUC), the better the model is at predicting correctly.

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
```

```python
# Train models
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train, y_train)
```

Here we are:

- Creating three models: Logistic Regression, Random Forest, and XGBoost

- Training each of them on the training data (X_train and y_train)

- The XGBoost model has a few extra settings:

use_label_encoder=False → stops it from giving a warning

eval_metric='logloss' → tells it to use log loss for evaluation

```python
# Plotting function
def plot_roc(model, name, color):
    probs = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, probs)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color=color, lw=2, label=f'{name} (AUC = {roc_auc:.2f})')
```

```python
# Create plot
plt.figure(figsize=(8, 6))
plot_roc(lr_model, "Logistic Regression", "blue")
plot_roc(rf_model, "Random Forest", "green")
plot_roc(xgb_model, "XGBoost Classifier", "red")

plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend(loc='lower right')
plt.show()
```

The diagonal line helps to compare your models. If a model’s curve is above this line, it’s doing better than random guessing.

**Hyperparameter Tuning**

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 8],
    'min_samples_split': [2, 5],
}

grid_search = GridSearchCV(RandomForestClassifier(random_state=42),
                           param_grid,
                           cv=5,
                           scoring='roc_auc',
                           n_jobs=-1)

grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
best_rf = grid_search.best_estimator_
```

```python
best_preds = best_rf.predict(X_test)
print(classification_report(y_test, best_preds))
print("AUC Score:", roc_auc_score(y_test, best_preds))
```

**Streamlit App (app.py)**

Goal: Predict if a patient has Diabetic Retinopathy using logistic regression

```python
import streamlit as st
import numpy as np
import joblib

# Load trained model
model = joblib.load("logistic_model.pkl")

# Streamlit UI
st.set_page_config(page_title="Diabetic Retinopathy Prediction")
st.title(" Diabetic Retinopathy Prediction (Logistic Regression)")
st.write("Enter patient information to predict the prognosis.")

# Input fields (update according to the model features)
age = st.number_input("Age", 0, 120, value=50)
systolic = st.number_input("Systolic BP", 80, 200, value=120)
diastolic = st.number_input("Diastolic BP", 50, 150, value=80)
cholesterol = st.number_input("Cholesterol", 100, 400, value=200)

# Collect input data
input_data = np.array([[age, systolic, diastolic, cholesterol]])

# Prediction button
if st.button("Predict"):
    prediction = model.predict(input_data)[0]
    if prediction == 1:
        st.error("Diabetic Retinopathy detected.")
    else:
        st.success("No Diabetic Retinopathy detected.")
```

 ### **Input Fields Explained**

1. Age = 50:

- Age of the patient in years

- Used because diabetic risk increases with age

2. Systolic BP = 120:

- Measures pressure during heartbeats (top number)

- Normal range; higher values suggest cardiovascular risk

3. Diastolic BP = 80:

- Pressure between heartbeats (bottom number)

- 80 is normal; elevated levels can indicate health issues

4. Cholesterol = 200:

- mg/dL in blood

- 200 is borderline-high, which may correlate with diabetes risk

### **What Happens on Predict Button**

Streamlit collects all 4 values as input → [50, 120, 80, 200]

Model runs .predict() using these

Output is:

0 → shows  No Retinopathy

1 → shows  Retinopathy Detected
